```python
RECORDINGS_PATH = "/Users/ntuan1/workspaces/content/drive/MyDrive/amd-training"
PICKLE_FILE_NAME  = "df-40-200-mfccs-kaiser_fast-01-25-19"
```

```python
#compute
import glob, os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import uuid

os.chdir(RECORDINGS_PATH)
df = pd.DataFrame(columns=['file_path', 'mfccs_40', 'chroma', 'mel','contrast','tonnetz', 'duration','label'])
df = df.fillna(0) # with 0s rather than NaNs
def process(file, label):
  try:
    print("processing {}".format(file))
    
    # here kaiser_fast is a technique used for faster extraction
    X, sample_rate = librosa.load(file, sr=16000,res_type='kaiser_fast')
    duration = librosa.get_duration(y=X, sr=sample_rate)
    stft = np.abs(librosa.stft(X))
    mfccs_40 = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
    
    S = librosa.feature.melspectrogram(y=X, sr=sample_rate, n_mels=128,fmax=8000)
    print(S.T)
    mel = np.mean(S.T,axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X),
    sr=sample_rate).T,axis=0)

    row = pd.Series({'file_path':file,'mfccs_40':mfccs_40, 'chroma':chroma, 'mel':mel, 'contrast':contrast, 'tonnetz':tonnetz, 'duration':duration,'label':label},name=7)
    df.loc[len(df)] = row  
    
  except Exception as e:
    print("Error encountered while parsing file: {} by {}".format(file, e))


def download_files(path, label, max=100):
  index = 0
  for file in glob.glob(path):
    print("{} file in folder".format(file))
    if index <=max:
      row = process(file, label)
      df.loc[len(df)] = row
      index = index +1

#save the df for reusabality
df.to_pickle(PICKLE_FILE_NAME+".pkl")
print(df.to_json())
```
